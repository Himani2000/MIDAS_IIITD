#here the function himani tweets ask for the username and the number of tweets as  the function 
#the  username can be  any twitter account for the question we have taken the midasIIITD

```python
def himani_tweets(username,number_of_tweets=250): 
          
        # Authorization to consumer key and consumer secret 
        auth = tweepy.OAuthHandler(CONSUMER_KEY, CONSUMER_SECRET) 
  
        # Access to user's access key and access secret 
        auth.set_access_token(ACCESS_TOKEN, ACCESS_TOKEN_SECRET) 
  
        # Calling  twiiter api 
        api = tweepy.API(auth) 
   
        # getting the tweets through calling the api by giving the username and the count of the tweets 
        tweets = api.user_timeline(screen_name=username,count=number_of_tweets) 
  
        # making a json file and appending the tweets 
        with open(username+".jsonl",'w') as f:
            for tweet in tweets:
                json.dump(tweet._json, f)
                f.write('\n')
        # converting the json file to the pandas dataframe 
        df=pd.read_json(username+".jsonl",lines=True)
        # returning the dataframe 
        return df
  
 
```
